```python
s_golay_filtered_signal =scipy.signal.savgol_filter(first_electrode,5,2)
#in the paper he used a quadratic plynomial and a frame span of size 5
plt.plot(time, s_golay_filtered_signal, label="smoothed signal")
plt.xlim([0, 4]);
plt.legend()
plt.show()
```

```python
cA, d1 = pywt.dwt(s_golay_filtered_signal, 'db4')
cA2, d2 = pywt.dwt(cA, 'db4')
cA3, d3 = pywt.dwt(cA2, 'db4')
a4, d4 = pywt.dwt(cA3, 'db4')
#optaining gamma, beta, alpha, theta and delta signals from our smoothed signal
```
