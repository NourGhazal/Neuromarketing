# Optaining gamma, beta, alpha, theta and delta signals from our smoothed signal

# Getting the Root square mean for the five filtered signals
we get the root square mean as a values representing the wave for our model to use in trainnig

```python
gamma_RMS = np.sqrt(sum([d1[i]**2 for i in range(d1.size)])/d1.size)
beta_RMS = np.sqrt(sum([d2[i]**2 for i in range(d2.size)])/d2.size)
alpha_RMS = np.sqrt(sum([d3[i]**2 for i in range(d3.size)])/d3.size)
theta_RMS = np.sqrt(sum([d4[i]**2 for i in range(d4.size)])/d4.size)
delta_RMS = np.sqrt(sum([a4[i]**2 for i in range(a4.size)])/a4.size)
print(gamma_RMS,beta_RMS,alpha_RMS,theta_RMS,delta_RMS)
```
