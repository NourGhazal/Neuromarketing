```python
results_SVM = {
    'user':[],
    'result_gamma':[],
    'result_beta':[],
    'result_theta':[],
    'result_alpha':[],
    'result_delta':[],
    'result_merged':[],
}
all_labels=[]
prediction_labels_gamma=[]
prediction_labels_beta=[]
prediction_labels_alpha=[]
prediction_labels_theta=[]
prediction_labels_delta=[]
prediction_labels_merged=[]

for name in ['Abhishek_','Ankur_sir_','Gautam_','Gautam_123_','Girvar_yadav_','Kishore_babu_','mahendra_','Mohit_','pawan_sahu_','pradeep_','Rajesh_el_','rajkumar_','Ravi_baba_','Ravi_ph_','Rockysingh_','Rupak_','Sachin_','Sandeep_','Soumendu_','Suraj_sir_','taufiq_','Veerpal_','Vijay_','Vipin_1_','Viraj_1_'] :
        labels=[]
        gamma_data1={
            "1stgamma":[],
            "2ndgamma":[],
            "3rdgamma":[],
            "4thgamma":[],
            "5thgamma":[],
            "6thgamma":[],
            "7thgamma":[],
            "8thgamma":[],
            "9thgamma":[],
            "10thgamma":[],
            "11thgamma":[],
            "12thgamma":[],
            "13thgamma":[],
            "14thgamma":[]}
        beta_data1={
            "1stbeta":[],
            "2ndbeta":[],
            "3rdbeta":[],
            "4thbeta":[],
            "5thbeta":[],
            "6thbeta":[],
            "7thbeta":[],
            "8thbeta":[],
            "9thbeta":[],
            "10thbeta":[],
            "11thbeta":[],
            "12thbeta":[],
            "13thbeta":[],
            "14thbeta":[]}
        alpha_data1={
            "1stalpha":[],
            "2ndalpha":[],
            "3rdalpha":[],
            "4thalpha":[],
            "5thalpha":[],
            "6thalpha":[],
            "7thalpha":[],
            "8thalpha":[],
            "9thalpha":[],
            "10thalpha":[],
            "11thalpha":[],
            "12thalpha":[],
            "13thalpha":[],
            "14thalpha":[]}
        theta_data1={
            "1sttheta":[],
            "2ndtheta":[],
            "3rdtheta":[],
            "4ththeta":[],
            "5ththeta":[],
            "6ththeta":[],
            "7ththeta":[],
            "8ththeta":[],
            "9ththeta":[],
            "10ththeta":[],
            "11ththeta":[],
            "12ththeta":[],
            "13ththeta":[],
            "14ththeta":[]}
        delta_data1={
            "1stdelta":[],
            "2nddelta":[],
            "3rddelta":[],
            "4thdelta":[],
            "5thdelta":[],
            "6thdelta":[],
            "7thdelta":[],
            "8thdelta":[],
            "9thdelta":[],
            "10thdelta":[],
            "11thdelta":[],
            "12thdelta":[],
            "13thdelta":[],
            "14thdelta":[]}
        for i in range(42):
            try:
                user_data = pd.read_csv("25-users/"+name+str(i+1)+".txt", sep=" ", header=None,names=["1st", "2nd", "3rd", "4th","5th","6th","7th","8th","9th","10th","11th","12th","13th","14th"])
                user_label = pd.read_csv("labels/"+name+str(i+1)+".lab",header=None,names=["label"])
            except:
                continue
            labels.append(user_label.values.tolist()[0][0])
            user_data.apply(lambda row : carf(row), axis = 1)
            
            first_electrode = user_data['1st']
            second_electrode = user_data['2nd']
            third_electrode = user_data['3rd']
            fourth_electrode = user_data['4th']
            fifth_electrode = user_data['5th']
            sixth_electrode = user_data['6th']
            seventh_electrode = user_data['7th']
            eighth_electrode = user_data['8th']
            nineth_electrode = user_data['9th']
            tenth_electrode = user_data['10th']
            eleventh_electrode = user_data['11th']
            twelveth_electrode = user_data['12th']
            thirteenth_electrode = user_data['13th']
            fouteenth_electrode = user_data['14th']



            s_golay_filtered_signal1 =scipy.signal.savgol_filter(first_electrode,5,2)
            s_golay_filtered_signal2 =scipy.signal.savgol_filter(second_electrode,5,2)
            s_golay_filtered_signal3 =scipy.signal.savgol_filter(third_electrode,5,2)
            s_golay_filtered_signal4 =scipy.signal.savgol_filter(fourth_electrode,5,2)
            s_golay_filtered_signal5 =scipy.signal.savgol_filter(fifth_electrode,5,2)
            s_golay_filtered_signal6 =scipy.signal.savgol_filter(sixth_electrode,5,2)
            s_golay_filtered_signal7 =scipy.signal.savgol_filter(seventh_electrode,5,2)
            s_golay_filtered_signal8 =scipy.signal.savgol_filter(eighth_electrode,5,2)
            s_golay_filtered_signal9 =scipy.signal.savgol_filter(nineth_electrode,5,2)
            s_golay_filtered_signal10 =scipy.signal.savgol_filter(tenth_electrode,5,2)
            s_golay_filtered_signal11 =scipy.signal.savgol_filter(eleventh_electrode,5,2)
            s_golay_filtered_signal12 =scipy.signal.savgol_filter(twelveth_electrode,5,2)
            s_golay_filtered_signal13 =scipy.signal.savgol_filter(thirteenth_electrode,5,2)
            s_golay_filtered_signal14 =scipy.signal.savgol_filter(fouteenth_electrode,5,2)


            cA1, d11 = pywt.dwt(s_golay_filtered_signal1, 'db4')
            cA12, d12 = pywt.dwt(cA1, 'db4')
            cA13, d13 = pywt.dwt(cA12, 'db4')
            a14, d14 = pywt.dwt(cA13, 'db4')
            gamma_RMS1 = np.sqrt(sum([d11[i]**2 for i in range(d11.size)])/d11.size)
            beta_RMS1 = np.sqrt(sum([d12[i]**2 for i in range(d12.size)])/d12.size)
            alpha_RMS1 = np.sqrt(sum([d13[i]**2 for i in range(d13.size)])/d13.size)
            theta_RMS1 = np.sqrt(sum([d14[i]**2 for i in range(d14.size)])/d14.size)
            delta_RMS1 = np.sqrt(sum([a14[i]**2 for i in range(a14.size)])/a14.size)
            gamma_data1['1stgamma'].append(gamma_RMS1)
            beta_data1['1stbeta'].append(beta_RMS1)
            alpha_data1['1stalpha'].append(alpha_RMS1)
            theta_data1['1sttheta'].append(theta_RMS1)
            delta_data1['1stdelta'].append(delta_RMS1)

            cA2, d21 = pywt.dwt(s_golay_filtered_signal2, 'db4')
            cA22, d22 = pywt.dwt(cA2, 'db4')
            cA23, d23 = pywt.dwt(cA22, 'db4')
            a24, d24 = pywt.dwt(cA23, 'db4')
            gamma_RMS2 = np.sqrt(sum([d21[i]**2 for i in range(d21.size)])/d21.size)
            beta_RMS2 = np.sqrt(sum([d22[i]**2 for i in range(d22.size)])/d22.size)
            alpha_RMS2 = np.sqrt(sum([d23[i]**2 for i in range(d23.size)])/d23.size)
            theta_RMS2 = np.sqrt(sum([d24[i]**2 for i in range(d24.size)])/d24.size)
            delta_RMS2 = np.sqrt(sum([a24[i]**2 for i in range(a24.size)])/a24.size)
            gamma_data1['2ndgamma'].append(gamma_RMS2)
            beta_data1['2ndbeta'].append(beta_RMS2)
            alpha_data1['2ndalpha'].append(alpha_RMS2)
            theta_data1['2ndtheta'].append(theta_RMS2)
            delta_data1['2nddelta'].append(delta_RMS2)

            cA3, d31 = pywt.dwt(s_golay_filtered_signal3, 'db4')
            cA32, d32 = pywt.dwt(cA3, 'db4')
            cA33, d33 = pywt.dwt(cA32, 'db4')
            a34, d34 = pywt.dwt(cA33, 'db4')
            gamma_RMS3 = np.sqrt(sum([d31[i]**2 for i in range(d31.size)])/d31.size)
            beta_RMS3 = np.sqrt(sum([d32[i]**2 for i in range(d32.size)])/d32.size)
            alpha_RMS3 = np.sqrt(sum([d33[i]**2 for i in range(d33.size)])/d33.size)
            theta_RMS3 = np.sqrt(sum([d34[i]**2 for i in range(d34.size)])/d34.size)
            delta_RMS3 = np.sqrt(sum([a34[i]**2 for i in range(a34.size)])/a34.size)
            gamma_data1['3rdgamma'].append(gamma_RMS3)
            beta_data1['3rdbeta'].append(beta_RMS3)
            alpha_data1['3rdalpha'].append(alpha_RMS3)
            theta_data1['3rdtheta'].append(theta_RMS3)
            delta_data1['3rddelta'].append(delta_RMS3)

            cA4, d41 = pywt.dwt(s_golay_filtered_signal4, 'db4')
            cA42, d42 = pywt.dwt(cA4, 'db4')
            cA43, d43 = pywt.dwt(cA42, 'db4')
            a44, d44 = pywt.dwt(cA43, 'db4')
            gamma_RMS4 = np.sqrt(sum([d41[i]**2 for i in range(d41.size)])/d41.size)
            beta_RMS4 = np.sqrt(sum([d42[i]**2 for i in range(d42.size)])/d42.size)
            alpha_RMS4 = np.sqrt(sum([d43[i]**2 for i in range(d43.size)])/d43.size)
            theta_RMS4 = np.sqrt(sum([d44[i]**2 for i in range(d44.size)])/d44.size)
            delta_RMS4 = np.sqrt(sum([a44[i]**2 for i in range(a44.size)])/a44.size)
            gamma_data1['4thgamma'].append(gamma_RMS4)
            beta_data1['4thbeta'].append(beta_RMS4)
            alpha_data1['4thalpha'].append(alpha_RMS4)
            theta_data1['4ththeta'].append(theta_RMS4)
            delta_data1['4thdelta'].append(delta_RMS4)

            cA5, d51 = pywt.dwt(s_golay_filtered_signal5, 'db4')
            cA52, d52 = pywt.dwt(cA5, 'db4')
            cA53, d53 = pywt.dwt(cA52, 'db4')
            a54, d54 = pywt.dwt(cA53, 'db4')
            gamma_RMS5 = np.sqrt(sum([d51[i]**2 for i in range(d51.size)])/d51.size)
            beta_RMS5 = np.sqrt(sum([d52[i]**2 for i in range(d52.size)])/d52.size)
            alpha_RMS5 = np.sqrt(sum([d53[i]**2 for i in range(d53.size)])/d53.size)
            theta_RMS5 = np.sqrt(sum([d54[i]**2 for i in range(d54.size)])/d54.size)
            delta_RMS5 = np.sqrt(sum([a54[i]**2 for i in range(a54.size)])/a54.size)
            gamma_data1['5thgamma'].append(gamma_RMS5)
            beta_data1['5thbeta'].append(beta_RMS5)
            alpha_data1['5thalpha'].append(alpha_RMS5)
            theta_data1['5ththeta'].append(theta_RMS5)
            delta_data1['5thdelta'].append(delta_RMS5)

            cA6, d61 = pywt.dwt(s_golay_filtered_signal6, 'db4')
            cA62, d62 = pywt.dwt(cA6, 'db4')
            cA63, d63 = pywt.dwt(cA62, 'db4')
            a64, d64 = pywt.dwt(cA63, 'db4')
            gamma_RMS6 = np.sqrt(sum([d61[i]**2 for i in range(d61.size)])/d61.size)
            beta_RMS6 = np.sqrt(sum([d62[i]**2 for i in range(d62.size)])/d62.size)
            alpha_RMS6 = np.sqrt(sum([d63[i]**2 for i in range(d63.size)])/d63.size)
            theta_RMS6 = np.sqrt(sum([d64[i]**2 for i in range(d64.size)])/d64.size)
            delta_RMS6 = np.sqrt(sum([a64[i]**2 for i in range(a64.size)])/a64.size)
            gamma_data1['6thgamma'].append(gamma_RMS6)
            beta_data1['6thbeta'].append(beta_RMS6)
            alpha_data1['6thalpha'].append(alpha_RMS6)
            theta_data1['6ththeta'].append(theta_RMS6)
            delta_data1['6thdelta'].append(delta_RMS6)


            cA7, d71 = pywt.dwt(s_golay_filtered_signal7, 'db4')
            cA72, d72 = pywt.dwt(cA7, 'db4')
            cA73, d73 = pywt.dwt(cA72, 'db4')
            a74, d74 = pywt.dwt(cA73, 'db4')
            gamma_RMS7 = np.sqrt(sum([d71[i]**2 for i in range(d71.size)])/d71.size)
            beta_RMS7 = np.sqrt(sum([d72[i]**2 for i in range(d72.size)])/d72.size)
            alpha_RMS7 = np.sqrt(sum([d73[i]**2 for i in range(d73.size)])/d73.size)
            theta_RMS7 = np.sqrt(sum([d74[i]**2 for i in range(d74.size)])/d74.size)
            delta_RMS7 = np.sqrt(sum([a74[i]**2 for i in range(a74.size)])/a74.size)
            gamma_data1['7thgamma'].append(gamma_RMS7)
            beta_data1['7thbeta'].append(beta_RMS7)
            alpha_data1['7thalpha'].append(alpha_RMS7)
            theta_data1['7ththeta'].append(theta_RMS7)
            delta_data1['7thdelta'].append(delta_RMS7)

            cA8, d81 = pywt.dwt(s_golay_filtered_signal8, 'db4')
            cA82, d82 = pywt.dwt(cA8, 'db4')
            cA83, d83 = pywt.dwt(cA82, 'db4')
            a84, d84 = pywt.dwt(cA83, 'db4')
            gamma_RMS8 = np.sqrt(sum([d81[i]**2 for i in range(d81.size)])/d81.size)
            beta_RMS8 = np.sqrt(sum([d82[i]**2 for i in range(d82.size)])/d82.size)
            alpha_RMS8 = np.sqrt(sum([d83[i]**2 for i in range(d83.size)])/d83.size)
            theta_RMS8 = np.sqrt(sum([d84[i]**2 for i in range(d84.size)])/d84.size)
            delta_RMS8 = np.sqrt(sum([a84[i]**2 for i in range(a84.size)])/a84.size)
            gamma_data1['8thgamma'].append(gamma_RMS8)
            beta_data1['8thbeta'].append(beta_RMS8)
            alpha_data1['8thalpha'].append(alpha_RMS8)
            theta_data1['8ththeta'].append(theta_RMS8)
            delta_data1['8thdelta'].append(delta_RMS8)

            cA9, d91 = pywt.dwt(s_golay_filtered_signal9, 'db4')
            cA92, d92 = pywt.dwt(cA9, 'db4')
            cA93, d93 = pywt.dwt(cA92, 'db4')
            a94, d94 = pywt.dwt(cA93, 'db4')
            gamma_RMS9 = np.sqrt(sum([d91[i]**2 for i in range(d91.size)])/d91.size)
            beta_RMS9 = np.sqrt(sum([d92[i]**2 for i in range(d92.size)])/d92.size)
            alpha_RMS9 = np.sqrt(sum([d93[i]**2 for i in range(d93.size)])/d93.size)
            theta_RMS9 = np.sqrt(sum([d94[i]**2 for i in range(d94.size)])/d94.size)
            delta_RMS9 = np.sqrt(sum([a94[i]**2 for i in range(a94.size)])/a94.size)
            gamma_data1['9thgamma'].append(gamma_RMS9)
            beta_data1['9thbeta'].append(beta_RMS9)
            alpha_data1['9thalpha'].append(alpha_RMS9)
            theta_data1['9ththeta'].append(theta_RMS9)
            delta_data1['9thdelta'].append(delta_RMS9)

            cA10, d101 = pywt.dwt(s_golay_filtered_signal10, 'db4')
            cA102, d102 = pywt.dwt(cA10, 'db4')
            cA103, d103 = pywt.dwt(cA102, 'db4')
            a104, d104 = pywt.dwt(cA103, 'db4')
            gamma_RMS10 = np.sqrt(sum([d101[i]**2 for i in range(d101.size)])/d101.size)
            beta_RMS10 = np.sqrt(sum([d102[i]**2 for i in range(d102.size)])/d102.size)
            alpha_RMS10 = np.sqrt(sum([d103[i]**2 for i in range(d103.size)])/d103.size)
            theta_RMS10 = np.sqrt(sum([d104[i]**2 for i in range(d104.size)])/d104.size)
            delta_RMS10 = np.sqrt(sum([a104[i]**2 for i in range(a104.size)])/a104.size)
            gamma_data1['10thgamma'].append(gamma_RMS10)
            beta_data1['10thbeta'].append(beta_RMS10)
            alpha_data1['10thalpha'].append(alpha_RMS10)
            theta_data1['10ththeta'].append(theta_RMS10)
            delta_data1['10thdelta'].append(delta_RMS10)

            cA11, d111 = pywt.dwt(s_golay_filtered_signal11, 'db4')
            cA112, d112 = pywt.dwt(cA11, 'db4')
            cA113, d113 = pywt.dwt(cA112, 'db4')
            a114, d114 = pywt.dwt(cA113, 'db4')
            gamma_RMS11 = np.sqrt(sum([d111[i]**2 for i in range(d111.size)])/d111.size)
            beta_RMS11 = np.sqrt(sum([d112[i]**2 for i in range(d112.size)])/d112.size)
            alpha_RMS11 = np.sqrt(sum([d113[i]**2 for i in range(d113.size)])/d113.size)
            theta_RMS11 = np.sqrt(sum([d114[i]**2 for i in range(d114.size)])/d114.size)
            delta_RMS11 = np.sqrt(sum([a114[i]**2 for i in range(a114.size)])/a114.size)
            gamma_data1['11thgamma'].append(gamma_RMS11)
            beta_data1['11thbeta'].append(beta_RMS11)
            alpha_data1['11thalpha'].append(alpha_RMS11)
            theta_data1['11ththeta'].append(theta_RMS11)
            delta_data1['11thdelta'].append(delta_RMS11)

            cA12, d121 = pywt.dwt(s_golay_filtered_signal12, 'db4')
            cA122, d122 = pywt.dwt(cA12, 'db4')
            cA123, d123 = pywt.dwt(cA122, 'db4')
            a124, d124 = pywt.dwt(cA123, 'db4')
            gamma_RMS12 = np.sqrt(sum([d121[i]**2 for i in range(d121.size)])/d11.size)
            beta_RMS12 = np.sqrt(sum([d122[i]**2 for i in range(d122.size)])/d12.size)
            alpha_RMS12 = np.sqrt(sum([d123[i]**2 for i in range(d123.size)])/d13.size)
            theta_RMS12 = np.sqrt(sum([d124[i]**2 for i in range(d124.size)])/d14.size)
            delta_RMS12 = np.sqrt(sum([a124[i]**2 for i in range(a124.size)])/a14.size)
            gamma_data1['12thgamma'].append(gamma_RMS12)
            beta_data1['12thbeta'].append(beta_RMS12)
            alpha_data1['12thalpha'].append(alpha_RMS12)
            theta_data1['12ththeta'].append(theta_RMS12)
            delta_data1['12thdelta'].append(delta_RMS12)

            cA13, d131 = pywt.dwt(s_golay_filtered_signal13, 'db4')
            cA132, d132 = pywt.dwt(cA13, 'db4')
            cA133, d133 = pywt.dwt(cA132, 'db4')
            a134, d134 = pywt.dwt(cA133, 'db4')
            gamma_RMS13 = np.sqrt(sum([d131[i]**2 for i in range(d131.size)])/d11.size)
            beta_RMS13 = np.sqrt(sum([d132[i]**2 for i in range(d132.size)])/d12.size)
            alpha_RMS13 = np.sqrt(sum([d133[i]**2 for i in range(d133.size)])/d13.size)
            theta_RMS13 = np.sqrt(sum([d134[i]**2 for i in range(d134.size)])/d14.size)
            delta_RMS13 = np.sqrt(sum([a134[i]**2 for i in range(a134.size)])/a14.size)
            gamma_data1['13thgamma'].append(gamma_RMS13)
            beta_data1['13thbeta'].append(beta_RMS13)
            alpha_data1['13thalpha'].append(alpha_RMS13)
            theta_data1['13ththeta'].append(theta_RMS13)
            delta_data1['13thdelta'].append(delta_RMS13)

            cA14, d141 = pywt.dwt(s_golay_filtered_signal14, 'db4')
            cA142, d142 = pywt.dwt(cA14, 'db4')
            cA143, d143 = pywt.dwt(cA142, 'db4')
            a144, d144 = pywt.dwt(cA143, 'db4')
            gamma_RMS14 = np.sqrt(sum([d141[i]**2 for i in range(d141.size)])/d141.size)
            beta_RMS14 = np.sqrt(sum([d142[i]**2 for i in range(d142.size)])/d142.size)
            alpha_RMS14 = np.sqrt(sum([d143[i]**2 for i in range(d143.size)])/d143.size)
            theta_RMS14 = np.sqrt(sum([d144[i]**2 for i in range(d144.size)])/d144.size)
            delta_RMS14 = np.sqrt(sum([a144[i]**2 for i in range(a144.size)])/a144.size)
            gamma_data1['14thgamma'].append(gamma_RMS14)
            beta_data1['14thbeta'].append(beta_RMS14)
            alpha_data1['14thalpha'].append(alpha_RMS14)
            theta_data1['14ththeta'].append(theta_RMS14)
            delta_data1['14thdelta'].append(delta_RMS14)
        
        merged = {**gamma_data1, **beta_data1,**alpha_data1,**theta_data1,**delta_data1}
        merged= pd.DataFrame.from_dict(merged)
        gamma_data1=pd.DataFrame.from_dict(gamma_data1)
        beta_data1=pd.DataFrame.from_dict(beta_data1)
        alpha_data1=pd.DataFrame.from_dict(alpha_data1)
        theta_data1=pd.DataFrame.from_dict(theta_data1)
        delta_data1=pd.DataFrame.from_dict(delta_data1)
        
        pca_gamma = PCA(n_components=0.9)
        principalComponents_gamma = pca_gamma.fit_transform(gamma_data1)
        gamma_data1 = pd.DataFrame(data = principalComponents_gamma)

        pca_beta = PCA(n_components=0.9)
        principalComponents_beta = pca_beta.fit_transform(beta_data1)
        beta_data1= pd.DataFrame(data = principalComponents_beta)
        
        pca_alpha = PCA(n_components=0.9)
        principalComponents_alpha = pca_alpha.fit_transform(alpha_data1)
        alpha_data1 = pd.DataFrame(data = principalComponents_alpha)

        pca_theta = PCA(n_components=0.9)
        principalComponents_theta = pca_theta.fit_transform(theta_data1)
        theta_data1 = pd.DataFrame(data = principalComponents_theta)

        pca_delta = PCA(n_components=0.9)
        principalComponents_delta = pca_delta.fit_transform(delta_data1)
        delta_data1 = pd.DataFrame(data = principalComponents_delta)


        pca_merged = PCA(n_components=0.9)
        principalComponents_merged = pca_merged.fit_transform(merged)
        merged = pd.DataFrame(data = principalComponents_merged)
        
        data_train_gamma, data_test_gamma, label_train_gamma, label_test_gamma = train_test_split(gamma_data1, labels, test_size=0.1)
        data_train_beta, data_test_beta, label_train_beta, label_test_beta = train_test_split(beta_data1, labels, test_size=0.1)
        data_train_alpha, data_test_alpha, label_train_alpha, label_test_alpha = train_test_split(alpha_data1, labels, test_size=0.1)
        data_train_theta, data_test_theta, label_train_theta, label_test_theta = train_test_split(theta_data1, labels, test_size=0.1)
        data_train_delta, data_test_delta, label_train_delta, label_test_delta = train_test_split(delta_data1, labels, test_size=0.1)
        data_train_merged, data_test_merged, label_train_merged, label_merged = train_test_split(merged, labels, test_size=0.1)
        
        results_SVM['user'].append(name)
        model_gamma= SVC()
        model_gamma.fit(data_train_gamma,label_train_gamma)
        predictions_gamma = model_gamma.predict(data_test_gamma)
        results_SVM['result_gamma'].append(classification_report(label_test_gamma,predictions_gamma,output_dict=True))
        model_beta= SVC()
        model_beta.fit(data_train_beta,label_train_beta)
        predictions_beta = model_beta.predict(data_test_beta)
        results_SVM['result_beta'].append(classification_report(label_test_beta,predictions_beta,output_dict=True))
        model_alpha= SVC()
        model_alpha.fit(data_train_alpha,label_train_alpha)
        predictions_alpha = model_alpha.predict(data_test_alpha)
        results_SVM['result_alpha'].append(classification_report(label_test_alpha,predictions_alpha,output_dict=True))
        model_theta= SVC()
        model_theta.fit(data_train_theta,label_train_theta)
        predictions_theta = model_theta.predict(data_test_theta)
        results_SVM['result_theta'].append(classification_report(label_test_theta,predictions_theta,output_dict=True))
        model_delta= SVC()
        model_delta.fit(data_train_delta,label_train_delta)
        predictions_delta = model_delta.predict(data_test_delta)
        results_SVM['result_delta'].append(classification_report(label_test_delta,predictions_delta,output_dict=True))
        model_merged= SVC()
        model_merged.fit(data_train_merged,label_train_merged)
        predictions_merged = model_merged.predict(data_test_merged)
        print(label_merged,predictions_merged,'\n')
        results_SVM['result_merged'].append(classification_report(label_merged,predictions_merged,output_dict=True))
```
